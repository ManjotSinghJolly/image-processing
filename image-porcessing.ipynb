```python
# Reading the image using OpenCV
image = cv2.imread('Cat.jpeg')

# As OpenCV reads images in BGR format, so we have to convert it to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Displaying the image using Matplotlib
plt.imshow(image_rgb)
plt.axis('off') # Turning off axis labels
plt.show()
```

```python
# Converting the image to grayscale using NumPy
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Displaying the original and grayscale images side by side
plt.figure(figsize = (10,5))

#Original image
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Grayscale image
plt.subplot(1,2,2)
plt.imshow(gray_image, cmap = 'gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
# Resizing the Image
# Defining the new dimensions for resizing
new_width = 300
new_height = 200

# Resizing the image using different interpolation methods
interpolation_methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC, cv2.INTER_LANCZOS4]
method_names = ['Nearest Neighbor', 'Bilinear', 'Bicubic', 'Lanczos']

plt.figure(figsize = (15, 5))

# Iterating over each interpolation method
for i, method in enumerate(interpolation_methods):
    # Resizing the image using the current interpolation method
    resized_image = cv2.resize(image, (new_width, new_height), interpolation=method)
    
    # Displaying the resized image
    plt.subplot(1, len(interpolation_methods), i+1)
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    plt.title(method_names[i])
    plt.axis('off')

plt.show()
```

```python
# Applying different filtering techniques
# 1. Blurring (Gaussioan Blur)
blurred_image = cv2.GaussianBlur(gray_image, (5,5), 0)

# 2. Sharpening (Using kernel for sharpening)
sharpening_kernel = np.array([[-1, -1, -1],
                             [-1, 9, -1],
                             [-1, -1, -1]])
sharpened_image = cv2.filter2D(gray_image, -1, sharpening_kernel)

# 3. Edge Detection (Canny Edge Detection)
edges_image = cv2.Canny(gray_image, 100, 200)

# Displaying the original and processed images side by side
plt.figure(figsize=(15,5))

# Original image
plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Blurred image
plt.subplot(1,4,2)
plt.imshow(blurred_image, cmap = 'gray')
plt.title('Blurred Image')
plt.axis('off')

# Sharpened image
plt.subplot(1,4,3)
plt.imshow(sharpened_image, cmap = 'gray')
plt.title('Sharpened Image')
plt.axis('off')

# Edge-detected image
plt.subplot(1,4,4)
plt.imshow(edges_image, cmap = 'gray')
plt.title('Edge-detected Image')
plt.axis('off')

plt.show()
```

```python
# Flipping the image

# Flipping the image horizontally
flipped_horizontal = cv2.flip(image, 1)

# Flipping the image vertically
flipped_vertical = cv2.flip(image, 0)

# Displaying the original and flipped images side by side
plt.figure(figsize = (10,5))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('Off')

# Flipped horizontally
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB))
plt.title('Flipped Horizontally')
plt.axis('off')

# Flipped vertically
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB))
plt.title('Flipped Vertically')
plt.axis('off')

plt.show()
```

```python
# Rotating the image

# Rotating the image by 90 degrees clockwise
rotated_90_clockwise = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

# Rotating the image by 90 degrees counterclockwise
rotated_90_counterclockwise = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Displaying the original and rotated images side by side
plt.figure(figsize=(10,5))

# Original image
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Rotated 90 degrees clockwise
plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(rotated_90_clockwise, cv2.COLOR_BGR2RGB))
plt.title('Rotated 90° Clockwise')
plt.axis('off')

# Rotated 90 degrees counterclockwise
plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(rotated_90_counterclockwise, cv2.COLOR_BGR2RGB))
plt.title('Rotated 90° Counterclockwise')
plt.axis('off')

plt.show()
```
